```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mp='/content/'
```

```python
btcbbesc181=pd.read_csv(mp+'btcbbesc181.csv')
btcmfc181=pd.read_csv(mp+'btcmfc181.csv')
btcmm181=pd.read_csv(mp+'btcmm181.csv')
btcbbesc1193=pd.read_csv(mp+'btcbbesc1193.csv')
btcmfc1193=pd.read_csv(mp+'btcmfc1193.csv')
btcmm1193=pd.read_csv(mp+'btcmm1193.csv')
```

```python
ethbbesc90=pd.read_csv(mp+'ethbbesc90.csv')
ethmm90=pd.read_csv(mp+'ethmm90.csv')
ethmfc90=pd.read_csv(mp+'ethmfc90.csv')
ethbbesc1198=pd.read_csv(mp+'ethbbesc1198.csv')
ethmfc1198=pd.read_csv(mp+'ethmfc1198.csv')
ethmm1198=pd.read_csv(mp+'ethmm1198.csv')
```

```python
btcbbesc181.shape
```

```python
bbesc181=btcbbesc181.values[-181:-91,1:]/ethbbesc90.values[-90:,1:]
```

```python
plt.plot(bbesc181)
plt.plot([np.min(bbesc181)]*90)
plt.plot([np.max(bbesc181)]*90)
plt.text(x=30,y=np.min(bbesc181),s=f'{np.min(bbesc181)}')
plt.text(x=30,y=np.max(bbesc181),s=f'{np.max(bbesc181)}')
plt.grid()
plt.show()
```

```python
bbesc1193=btcbbesc1193.values[-1193:,1:]/ethbbesc1198.values[-1198:-5,1:]
```

```python
plt.plot(bbesc1193)
plt.plot([np.min(bbesc1193)]*1193)
plt.plot([np.max(bbesc1193)]*1193)
plt.text(x=500,y=np.min(bbesc1193),s=f'{np.min(bbesc1193)}')
plt.text(x=500,y=np.max(bbesc1193),s=f'{np.max(bbesc1193)}')
plt.grid()
plt.show()
```
